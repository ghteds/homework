```python
#Observations
# Temperature (F) vs. Latitude
# Humidity (%) vs. Latitude
# Cloudiness (%) vs. Latitude
# Wind Speed (mph) vs. Latitude
#LATITUDE VS TEMPERATURE,HUMIDITY,CLOUDINESS, WIND SPEED

# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import requests
import time

# Import API key
import api_keys

# Incorporated citipy to determine city based on latitude and longitude
from citipy import citipy

# Output File (CSV)
output_data_file = "cities.csv"

# Range of latitudes and longitudes
lat_range = (-90, 90)
lng_range = (-180, 180)
```

## Generate Cities List

```python
# List for holding lat_lngs and cities
lat_lngs = []
cities = []

# Create a set of random lat and lng combinations
lats = np.random.uniform(low=-90.000, high=90.000, size=1500)
lngs = np.random.uniform(low=-180.000, high=180.000, size=1500)
lat_lngs = zip(lats, lngs)

# Identify nearest city for each lat, lng combination
for lat_lng in lat_lngs:
    city = citipy.nearest_city(lat_lng[0], lat_lng[1]).city_name
    
    # If the city is unique, then add it to a our cities list
    if city not in cities and len(cities) < 501:
        cities.append(city)

# Print the city count to confirm sufficient count
len(cities)
```

## Perform API Calls

```python
# OpenWeatherMap API Key
api_key = api_keys.api_key

# Starting URL for Weather Map API Call
url = "http://api.openweathermap.org/data/2.5/weather?units=Imperial&APPID=" + api_key 

```

```python
cityName=[]
cityID=[]
theCountry=[]
latitude=[]
longitude=[]
temperatureMax=[]
cloudiness=[]
windSpeed=[]
humidity=[]
theDate=[]


for city in cities:
    try:
        response=requests.get(urlPlus+city).json()
        print(response['name'] + ' id:' + str(response['id']))
        cityName.append(response['name']) #name
        cityID.append(response['id']) #city id
        theCountry.append(response['sys']['country']) #country
        latitude.append(response['coord']['lat']) #lat
        longitude.append(response['coord']['lon']) #long
        temperatureMax.append(response['main']['temp_max']) #tmax
        cloudiness.append(response['clouds']['all']) #cloud %
        windSpeed.append(response['wind']['speed']) #ws
        humidity.append(response['main']['humidity']) #hum
        theDate.append(response['dt']) #the utc dt
    except KeyError:
          print('does not work')  
    
```

```python
print('Name: '  + str(len(cityName)) + '\n'+ \
        'cityID: ' +  str(len(cityID)) + '\n' + \
        'theCountry: ' + str(len(theCountry))  + '\n' + \
        'latitude: ' + str(len(latitude))  + '\n' + \
        'longitude: ' + str(len(longitude)) + '\n' + \
        'temperatureMax: ' + str(len(temperatureMax))  + '\n' + \
        'cloudiness: ' + str(len(cloudiness)) + '\n' + \
        'windSpeed: ' + str(len(windSpeed)) + '\n' + \
        'humidity: ' + str(len(humidity)) + '\n' + \
        'theDate: ' + str(len(theDate)))


```

```python
#create dictionary and df
weatherDictionary={
    'Name': cityName,
    #'City': cities,
    'CityID': cityID,
    'Country':theCountry,
    'Latitude': latitude,
    'Longitude': longitude,
    'TemperatureMax': temperatureMax,
    'Cloudiness': cloudiness,
    'WindSpeed': windSpeed,
    'Humidity': humidity,
    'Date': theDate
}

weatherData=pd.DataFrame(weatherDictionary)
```

```python
# Temperature (F) vs. Latitude
# Humidity (%) vs. Latitude
# Cloudiness (%) vs. Latitude
# Wind Speed (mph) vs. Latitude
plt.scatter(weatherData['Latitude'],weatherData['TemperatureMax'], marker='o')
plt.title("Latitude vs Temperature")
plt.ylabel('Temperature(F)')
plt.xlabel('Latitude')
#plt.xlim()
plt.savefig("latVStemperature.png")

#plt.xlimit auto
#
```
